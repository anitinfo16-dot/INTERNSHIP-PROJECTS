Section 2: k-NN & Distance Logic

1. K closest Points to Origin

```python
import math

def k_closest_points(points,k):
    distance_list=[]
    for (x,y) in points:
        distance=math.sqrt(x**2+y**2)
        distance_list.append((distance,(x,y)))
    distance_list.sort()
    result = [distance_list[i][1] for i in range(k)]
    return result
if __name__ == "__main__":
    points = [(1,3),(3,4),(2,-1),(5,2)]
    k = 2
    print("k_closest_points:",k_closest_points(points,k))
```

2.Nearest point with Same X or Y

```python
def nearest_same_x_or_y(points,x,y):
    min_dist=float('inf')
    neareset=None
    for px,py in points:
        if px==x or py==y:
            dist=(px-x)**2+(py-y)**2
            if dist<min_dist:
                min_dist=dist
                neareset=(px,py)
    return neareset
points = [(1,2),(3,4),(3,2),(5,2)]
x,y=3,3
print(nearest_same_x_or_y(points,x,y))
```

Section 3: Clustering & Grouping

1.Group Anagrams

```python
def group_anagrams(words):
    groups={}
    for word in words:
        key="".join(sorted(word))
        if key in groups:
            groups[key].append(word)
        else:
            groups[key]=[word]
    return list(groups.values())
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
result = group_anagrams(words)
print(result)
```

Section 4 : Matrix & Image Processing

2.Set Matrix Zeroes

```python
def set_matrix_zeroes(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    zero_rows = set()
    zero_cols = set()
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                zero_rows.add(i)
                zero_cols.add(j)
    for i in zero_rows:
        for j in range(cols):
                matrix[i][j] = 0
    for j in zero_cols:
        for i in range(rows):
            matrix[i][j] = 0
    return matrix
matrix=[[1,1,1],[1,0,1],[1,1,1]]
result = set_matrix_zeroes(matrix)
for row in result:
    print(row)
```

Section 5 : Sliding Window(CNN Analogy)

1.Maximum Average Subarray

```python
del sum
def max_average_subarray(nums, k):
    window_sum=sum(nums[:k])
    max_sum=window_sum
    for i in range(k, len(nums)):
        window_sum = window_sum - nums[i-k] + nums[i]
        max_sum = max(max_sum, window_sum)
    return max_sum / k
nums = [1, 12, -5, -6, 50, 3]
k = 4
result = max_average_subarray(nums, k)
print(result)
```

Mini Project : Build Your Own k-NN

```python
import math

dataset = [
    (150, 7, 'apple'),
    (130, 6, 'orange')
]

new_point = (140, 6.5)

def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

nearest_label = None
min_distance = float('inf')

for fruit in dataset:
    distance = euclidean_distance(new_point, fruit[:2])

    print(f"Distance from {fruit[2]} = {distance}")

    if distance < min_distance:
        min_distance = distance
        nearest_label = fruit[2]

print("Predicted Label:", nearest_label)
```
